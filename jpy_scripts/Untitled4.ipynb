```python
import numpy as np
import pandas as pd

import sys
sys.path.append("..")
import azapy as az
```

```python
sdate = pd.to_datetime("2012-01-01")
edate = pd.to_datetime("today")
symb = ['GLD', 'TLT', 'XLV', 'VGT', 'PSJ']

mktdir = "./MkTdata"

# force=True read from alphavantage server
# force=False try to read first from the local directory if data exists if not from alphavantage
mktdata = az.readMkT(symb, dstart = sdate, dend = edate, 
                     dir=mktdir, force=False) 
```

```python
alpha_99 = [0.99]
alpha_975 = [0.975]
alpha_95 = [0.95]
alpha_90 = [0.9]
alpha_80 = [0.8]
alpha_70 = [0.7]
alpha_60 = [0.6]
alpha1 = [0.99, 0.975, 0.95]
alpha2 = [0.99, 0.975, 0.95, 0.90]
alpha3 = [0.99, 0.80]
alpha4 = [0.95, 0.80]
alpha5 = [0.90, 0.70]

rtype = 'Sharpe'
```

```python
models = {'P_99': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha_99}, 'rtype': rtype},
          'P_975': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha_975}, 'rtype': rtype},
          'P_95': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha_95}, 'rtype': rtype},
          'P_90': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha_90}, 'rtype': rtype},
          'P_80': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha_80}, 'rtype': rtype},
          'P_70': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha_70}, 'rtype': rtype},
          'P_60': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha_60}, 'rtype': rtype},
          'P1': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha1}, 'rtype': rtype},
          'P2': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha2}, 'rtype': rtype},
          'P3': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha3}, 'rtype': rtype},
          'P4': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha4}, 'rtype': rtype},
          'P5': {'type': 'Port_CVaR', 'm_param': {'mu': 0., 'alpha': alpha5}, 'rtype': rtype}}
```

```python
port = []
pp = {}
for key, val in models.items():
    ppz = getattr(az, val['type'])
    pp_ = ppz(mktdata, pname=key)
    pp[key] = pp_
    port_ = pp_.set_model(**val['m_param'])
    port.append(port_)
```
