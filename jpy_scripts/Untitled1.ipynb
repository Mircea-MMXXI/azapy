```python
import sys
sys.path.append("..")
import azapy as az

print(f"azapy version {az.getVersion()}")
```

```python
symb = ['GLD', 'TLT', 'XLV', 'VGT', 'PSJ']

sdate = "2012-01-01"
edate = "2021-07-27"
#edate = "2022-11-04"
edate = 'today'

mktdir = "../MkTdata"

mktdata = az.readMkT(symb, sdate=sdate, edate=edate, file_dir=mktdir, verbose=False)
```

```python
rtype = 'InvNdiverse'
hlength = 1.25

alpha_CVaR = [0.95, 0.90, 0.85]
coef_CVaR = [1.] * len(alpha_CVaR)

alpha_EVaR = [0.70, 0.65]
coef_EVaR = [1.] * len(alpha_EVaR)

alpha_SMCR = [0.9, 0.80]


coef_MAD = [1./3.] * 3
coef_LSD = [1./3.] * 3
alpha_BTAD = [-0.01, 0]
alpha_BTSD = [-0.01, 0]


models = {'P_CVaR': {'type': 'Port_CVaR', 'm_param': {'alpha': alpha_CVaR, 'coef': coef_CVaR, 'rtype': rtype, 'hlength': hlength}},
          'P_EVaR': {'type': 'Port_EVaR', 'm_param': {'alpha': alpha_EVaR, 'coef': coef_EVaR, 'rtype': rtype, 'hlength': hlength}},
          'P_SMCR': {'type': 'Port_SMCR', 'm_param': {'alpha': alpha_SMCR, 'rtype': rtype, 'hlength': hlength}},
          'P_MAD': {'type': 'Port_MAD', 'm_param': {'coef': coef_MAD, 'rtype': rtype, 'hlength': hlength}},
          'P_LSD': {'type': 'Port_LSD', 'm_param': {'coef': coef_LSD, 'rtype': rtype, 'hlength': hlength}},
          'P_BTAD': {'type': 'Port_BTAD', 'm_param': {'alpha': alpha_BTAD, 'rtype': rtype, 'hlength': hlength}},
          'P_BTSD': {'type': 'Port_BTSD', 'm_param': {'alpha': alpha_BTSD, 'rtype': rtype, 'hlength': hlength}},
          #'P_GINI': {'type': 'Port_GINI', 'm_param': {'rtype': rtype, 'hlength': hlength}},
          'P_MV': {'type': 'Port_MV', 'm_param': {'rtype': rtype, 'hlength': hlength}},
          'P_SD': {'type': 'Port_SD', 'm_param': {'rtype': rtype, 'hlength': hlength}},
          'P_N': {'type': 'Port_ConstW', 'm_param': {'ww': None}},
         }
```

```python
port = []
pp = {}
for key, val in models.items():
    ppz = getattr(az, val['type'])
    pp_ = ppz(mktdata, pname=key)
    pp[key] = pp_
    port_ = pp_.set_model(**val['m_param'])
    port.append(port_)
```

```python
ps = az.Port_Simple(port, col='close', pname='ALL')
_ = ps.set_model()
```

```python
_ = ps.port_view_all(componly=True, fancy=True, title="Relative performance")
```
