```python
import numpy as np
import sys
sys.path.append("..")
import azapy as az

print(f"azapy version {az.version()}")
```

```python
mktdir = "../MkTdata"

sdate = "2012-01-01"
edate = "2022-07-27"

symb = ['GLD', 'TLT', 'XLV', 'IHI', 'PSJ',]

mktdata = az.readMkT(symb, sdate=sdate, edate=edate, file_dir=mktdir)
```

```python
alpha = np.array([0.95, 0.925, 0.9])
coef = np.full(len(alpha), 1/len(alpha))

# set Port_CVaR class
p4 = az.Port_CVaR(mktdata, pname='mCVaRPort')
```

```python
rtypes = ["MaxDiverse", "MinRisk", "Sharpe", "InvNrisk", "InvNdrr", "InvNdiverse"]
hlength = 1.25

port = []
for rty in rtypes:
    port4 = p4.set_model(alpha=alpha, rtype=rty, hlength=hlength)  
    port4.columns = [rty]
    port.append(port4)
```

```python
p5 = az.Port_ConstW(mktdata, pname="EWP")
port5 = p5.set_model()
port.append(port5)
```

```python
pp = az.Port_Simple(port)
pp.set_model()
_ = pp.port_view_all(componly=True, 
     title="mCVaR Portfolios - Relative Performance", 
     xlabel="year", fancy=True)
```
